```python
from bokeh.plotting import figure
from bokeh.io import output_notebook
from bokeh.io import show
```

```python
#Output the plot in notebook itself
output_notebook()
```

# Basic Plot

```python
#Scatter plot
```

```python
x = [ 5, 3, 2, 1, 0 ]
y = [25, 9, 4, 1, 0 ]

plot = figure(width=200, height=200)
plot.circle(x, y)

show(plot)
```

```python
#Scatter with other markers
```

```python
x = [ 5, 3, 2, 1, 0 ]
y = [25, 9, 4, 1, 0 ]

plot = figure(width=200, height=200)
plot.cross(x, y)

plot1 = figure(width=200, height=200)
plot1.diamond(x, y)

show(plot)
show(plot1)
```

```python
#Bar plot
```

```python
x = [ 3, 4, 5]
y = [ 9, 16, 25]

plot = figure(width=200, height=200)
plot.vbar(x, top=y, width=0.5)

show(plot)
```

```python
#Line plot
```

```python
x = [ 5, 3, 2, 1, 0 ]
y = [25, 9, 4, 1, 0 ]

plot = figure(width=200, height=200)
plot.line(x, y, color="red")

show(plot)
```

```python
#Line with datetime axis
```

```python
from bokeh.sampledata.glucose import data
print(data.head())

# reduce data size to one week
week = data.loc['2010-10-01':'2010-10-08']

plot = figure(x_axis_type="datetime", title="Glocose Range", plot_height=350, plot_width=800)
plot.xgrid.grid_line_color=None
plot.ygrid.grid_line_alpha=0.5
plot.xaxis.axis_label = 'Time'
plot.yaxis.axis_label = 'Value'

plot.line(week.index, week.glucose)

show(plot)
```
