#### 1.3. We see that there two columns which are Null (NaN). we have to solve this issue.
#####         a.  the column 'YoutubeLikes' has total 1 null value i.e. )0.66% of total rows 
#####         b. the column 'YoutubeDislikes' has total 3 null values i.e. 2% of total rows.. 

```python
## function to replace variable (x) with value (y)
def replaceNaN(df, x, y):
    df[x]=df[x].fillna(y)       
```

```python
# Filling missing values with medians of the columns
median_ytlikes = data['YoutubeLikes'].median()
print(' Mean of YoutubeLikes =>', median_ytlikes)
replaceNaN(data, 'YoutubeLikes', median_ytlikes)

# Filling missing values with medians of the columns
median_ytdislikes = data['YoutubeDislikes'].median()
print(' Mean of YoutubeDislikes =>', median_ytdislikes)
replaceNaN(data, 'YoutubeDislikes', median_ytdislikes)
```

```python
# after imputation, there should not be nay missing values
data.isnull().sum()
```

##### 1.4 In the dataset we observe that the column 'Budget' has 'Cr' has come as prefix or suffix. So we can remove the literal 'Cr', 'CR', 'cr'

```python
## function to replace value (x) with value (y)
def replaceVal(data, x, y):
    data.replace(to_replace =x, value = y, regex = True, inplace=True)
```

```python
replaceVal(data['Budget'], 'Cr', '')
data['Budget']=data.Budget.astype(float)

replaceVal(data['BoxOfficeCollection'], 'Cr', '')
data['BoxOfficeCollection']=data.BoxOfficeCollection.astype(float)
```

##### 1.6 The column 'ReleaseDate' has some invalid dates, we need to correct that column

```python
from dateutil.parser import parse

def checkInvalidDate(movie_id, dt):
    try:
        dt = dt.strip()
        return parse(dt).date()
    except ValueError as e:
        print(f"Invalid date '{dt}' for the movie id '{movie_id}' - reason: {e}")
        return dt
```

```python
def validateDate(data, dt):
    dt = dt.strip()
    return parse(dt)
```

```python
data.apply(lambda x: checkInvalidDate(x.MovieId, str(x.ReleaseDate)), axis=1)
```

###### As we see the results, we have three invalid dates which we need to correct

```python
replaceVal(data['ReleaseDate'], 'Sqp', 'Sep')
replaceVal(data['ReleaseDate'], 'Jum', 'Jun')
replaceVal(data['ReleaseDate'], 'juu', 'Jun')

# we can confrm that the dates are now valid
data.apply(lambda x: checkInvalidDate(x.MovieId, str(x.ReleaseDate)), axis=1)
```

##### 1.7 The column 'ReleaseTime' has some invalid values. We need to correct that column

```python
def checkInvalidString(movie_id, val):
    try:
        if isinstance(int(val), int):
            return val
        else:
            print(f"Invalid String value '{val}' for the movie id '{movie_id}'")
    except ValueError as e :
        print(f"Invalid String value '{val}' for the movie id '{movie_id}'")
        return val
```

```python
def checkInvalidNumber(movie_id, val):
    try:
        if isinstance(int(val), int):
            return val
        else:
            print(f"Invalid Int value '{val}' for the movie id '{movie_id}'")
    except ValueError as e :
        print(f"Invalid Int value '{val}' for the movie id '{movie_id}'")
        return val
```

```python
data.apply(lambda x: checkInvalidString(x.MovieId, str(x.ReleaseTime)), axis=1)
```

##### we can see above that we have so many rows with faulty 'ReleaseTime', such as 'HS', 'LW', 'N', 'FS'.
##### after cleaning, we should not see any invalid value

```python
replaceVal(data['ReleaseTime'], 'LW', '1')
replaceVal(data['ReleaseTime'], 'HS', '2')
replaceVal(data['ReleaseTime'], 'N', '3')
replaceVal(data['ReleaseTime'], 'FS', '4')

# we can confrm that the dates are now valid
data.apply(lambda x: checkInvalidString(x.MovieId, str(x.ReleaseTime)), axis=1)
```

## 4. Which month of the year, maximum movie releases are seen?

```python
from dateutil.parser import parse

data['ReleaseDate'] = data['ReleaseDate'].apply(lambda x: pd.to_datetime(parse(x).date(), errors='coerce'))
```

```python
## Now finding the count of movies release in Monthly frequency
max_release_year = data.set_index('ReleaseDate').groupby(pd.Grouper(freq="M")).agg('count')['MovieId']

print('Maximum movies released =>', str(max_release_year.max()))
print(str.format('Maximum movies released in the Month of {0} and Year {1}',str(max_release_year.idxmax().date().strftime("%B")),str(max_release_year.idxmax().date().year)))
```

## 5. Which month of the year typically sees most releases of high budgeted movies, that is, movies with budget of 25 crore or more?

```python
## Now finding the count of movies release in Monthly frequency

max_release_year = data[data['Budget'] > 25].set_index('ReleaseDate').groupby(pd.Grouper(freq="M")).agg('count')['MovieId']

print('Maximum movies released =>', str(max_release_year.max()))
print(str.format('Maximum movies released in the Month of {0} and Year {1}',str(max_release_year.idxmax().date().strftime("%B")),str(max_release_year.idxmax().date().year)))
```

## 6. Which are the top 10 movies with maximum return of investment (ROI)? Calculate ROI as (Box office collection – Budget) / Budget

```python
# calculate ROI
roi = (data['BoxOfficeCollection'] - data['Budget']) / data['Budget']

# Now we have ROI as a colum, we can use it to find top 10 movies by max ROI
data['ROI'] = roi

data.sort_values(['ROI'], ascending=False)[0:10][['MovieId','MovieName','ROI']].iloc[0:, 1:]
```
